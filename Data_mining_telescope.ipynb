```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("telescope_data.csv")
print(df.head())

df = df.iloc[:, :10]

print("\nMultivariate mean vector:\n", df.mean())
```

```python
df_centered = df - df.mean()
print(df_centered.head())

n = df.shape[0]
cov_matrix_inner = np.dot(df_centered.T, df_centered)/n
print("\nCovariance matrix as inner proucts:\n", cov_matrix_inner)
```

```python
cov_matrix_outer = np.cov(df_centered, rowvar=False)
print("\nCovariance matrix as outer product:\n", cov_matrix_outer)
```

```python

attr1 = df_centered.iloc[:, 0]
attr2 = df_centered.iloc[:, 1]
cosine_angle_attr1_attr2 = np.dot(attr1, attr2) / (np.linalg.norm(attr1) * np.linalg.norm(attr2))
print("Correlation between Attributes 1 and 2; cosine angle: ", cosine_angle_attr1_attr2)

sns.scatterplot(x=attr1, y=attr2)
plt.xlabel('Attribute 1')
plt.ylabel('Attribute 2')
plt.title('Scatter Plot between Attribute 1 and Attribute 2')
plt.show()
```

```python

sns.kdeplot(df.iloc[:, 0], fill=True)
plt.title('Probability Density Function of Attribute 1')
plt.xlabel('Attribute 1')
plt.ylabel('Density')
plt.show()
```

```python

line = np.array([1.6, -3.5])
line_unit = line / np.linalg.norm(line)
projection = np.dot(df_centered.iloc[:, [0, 1]], line_unit)

sns.scatterplot(x=projection, y=np.zeros_like(projection))
plt.xlabel('Projection')
plt.title('Projection of Attribute 1 and 2 to the line [1.6, -3.5]')
plt.show()
```

```python

variance = df.var()
largest_variance_attr = variance.idxmax(), variance.max()
smallest_variance_attr = variance.idxmin(), variance.min()

print("Attribute with the largest variance, variance: ", largest_variance_attr)
print("\nAttribute with the smallest variance, variance: ", smallest_variance_attr)
```

```python

covariance = pd.DataFrame(cov_matrix_outer).unstack()
covariance = covariance[covariance.index.get_level_values(0) != covariance.index.get_level_values(1)]
largest_covariance_attr = covariance.idxmax(), covariance.max()
smallest_covariance_attr = covariance.idxmin(), covariance.min()

print("Attribute pair with the largest covariance, covariance: ", largest_covariance_attr)
print("\nAttribute pair with the smallest covariance, covariance: ", smallest_covariance_attr)
```
